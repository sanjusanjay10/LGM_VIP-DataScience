```python
import  cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(title,img,cmap='RdBu'):
    plt.imshow(img,cmap=cmap)
    plt.title(title)
```

```python
image=plt.imread("/content/HORSE.JPG")
show("HORSE",image,'RdBu')
#cv2.imshow("HORSE",image)
#cv2.waitKey(0)
```

CONVERT TO GRAYSCALE IMAGE


```python
#convert to grayscale image
gray_image=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
show("gray_image",gray_image,'gray')
#cv2.imshow("gray_image",gray_image)
#cv2.waitKey(0)
```

INVERT THE GRAYSCALE IMAGE

```python
#invert the grayscale image
inverted_image =255-gray_image
show("inverted_image",inverted_image,'gray')
#cv2.imshow("inverted_image",inverted_image)
#cv2.waitKey(0)
```

BLUR THE INVERTED IMAGE

```python
#blur the inverted image
blurred_image =cv2.GaussianBlur(inverted_image,(21,21),0)
show("blurred_image",blurred_image,'gray')
#cv2.imshow("blurred_image",blurred_image)
#cv2.waitKey(0)
```

CREATE PENCIL SKETCH IMAGE

```python
inverted_blurred=255-blurred_image
pencil_sketch=cv2.divide(gray_image,inverted_blurred,scale=256.0)
show("pencil_sketch",pencil_sketch,'gray')
```
